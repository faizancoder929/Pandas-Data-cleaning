```python
import pandas as pd
import numpy as np

# 1. Simulating a terrible spreadsheet
client_sales_data = {
    'Transaction_ID': [101, 102, 103, 104, 104], # Note: 104 is a duplicate!
    'Product_Name': [' Laptop ', ' smartphone', 'Tablet', 'Laptop ', 'Laptop '],
    'Price_Each': ['$1200', '500', 'Missing', '$1200', '$1200'],
    'Quantity_Sold': [2.0, np.nan, 1.0, 2.0, 2.0] # Note: nan means an empty cell
}

# 2. Convert it into a Pandas DataFrame
df = pd.DataFrame(client_sales_data)
print("--- Messy Input Dataframe ---")
print(df.to_string())
```

```python
# 1. Remove duplicate rows
df = df.drop_duplicates()

# 2. Fix the formatting bugs in text columns
# .str.strip() removes hidden spaces. .str.title() forces proper capitalization (e.g., 'Smartphone')
df['Product_Name'] = df['Product_Name'].str.strip().str.title()

# 3. Clean up the price column
# We remove the '$' sign and force the word 'Missing' to become an empty value (NaN)
df['Price_Each'] = df['Price_Each'].str.replace('$', '', regex=False)
df['Price_Each'] = pd.to_numeric(df['Price_Each'], errors='coerce')

# 4. Fill in missing empty values
# Let us assume missing quantities should default to 1 sale
df['Quantity_Sold'] = df['Quantity_Sold'].fillna(1)

# 5. Create a brand new calculated column for the client!
df['Total_Revenue'] = df['Price_Each'] * df['Quantity_Sold']

print("--- Cleaned and Updated Dataframe ---")
print(df)
```

```python
# 1. Fill the missing tablet price with 300 and SAVE it back to the column
df['Price_Each'] = df['Price_Each'].fillna(300)

# 2. Recalculate the revenue and SAVE it back to overwrite the old column
df['Total_Revenue'] = df['Price_Each'] * df['Quantity_Sold']

# 3. Print to verify the clean updates
print("--- Final Updated Data ---")
print(df)
```

```python
#To convert a column's data type to a whole integer in Pandas, we use .astype(int)
df['Quantity_Sold'] = df['Quantity_Sold'].astype(int)
print(df)
```

```python
# Save the clean data to a CSV file.
# index=False prevents Pandas from adding a column of row numbers (0, 1, 2) which clients hate.
df.to_csv('cleaned_sales_report.csv', index=False)
print("File successfully saved to the cloud workspace!")
```
